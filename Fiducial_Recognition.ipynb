```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
%reload_ext watermark
%watermark -m -v -p numpy,matplotlib
print("opencv "+cv2.__version__)
matplotlib.rcParams['savefig.dpi'] = 200
```

```python
def getFiducial(filename):
    plt.figure()
    
    img = cv2.imread(filename,cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (img.shape[1]//4,img.shape[0]//4))
    w, h = img.shape
    
    plt.subplot(2,2,1);plt.imshow(img);plt.axis('off')
    
    _, img_thresh = cv2.threshold(img,180,1,cv2.THRESH_BINARY)
    plt.subplot(2,2,2);plt.imshow(img_thresh);plt.axis('off')
    
    element = cv2.getStructuringElement(cv2.MORPH_CROSS, (4,4))
    morph = cv2.dilate(img_thresh,element)
    plt.subplot(2,2,3);plt.imshow(morph);plt.axis('off')
    
    _, contours, _ = cv2.findContours(morph.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    def size_filter(contour):
        return  0.013 < cv2.contourArea(contour)/(w*h) < 0.016
    def ar_filter(contour):
        rect = cv2.minAreaRect(contour)
        ar = rect[1][0]/rect[1][1]
        return 0.9 < ar < 1.1
    
    contours = filter(size_filter, contours)
    contours = filter(ar_filter, contours)
    contours = sorted(contours, key = lambda x: cv2.contourArea(x), reverse=True)
    if len(contours) != 1:
        raise Exception("Could not isolate fiducials {} Shapes passed filters".format(len(contours)))
    
    fid_contour = contours[0]
    mask = np.zeros(morph.shape,dtype=np.uint8)
    cv2.drawContours(mask,[fid_contour],0,(10),-1)
    
    rect = cv2.minAreaRect(fid_contour)
    box = cv2.boxPoints(rect)
    maskRect = mask.copy()
    box = np.int0(box)
    cv2.drawContours(maskRect, [box], 0, (20), 2 )
    plt.subplot(2,2,4);plt.imshow(maskRect);plt.axis('off')
    
    s = "Found Fidicial!\nPosition={x},{y}\nRotation={phi} degrees".format(x=rect[0][0],y=rect[0][1],phi=rect[2])
    print(s)
    
    return rect
```

#BBM Fiducial Recognition
This program uses OpenCV to find a fiducial marking from a Bump-Bonded Module(BBM) in an input image.

The procedure is essentially the following:
1. Reduce resolution by a factor of 4 to shorten processing time
2. Threshold the image with a hard-coded cutoff. (This may need to be adjustable)
3. Dilate the foreground(Fiducial + noise) to close any small holes in image
4. Extract contours from image
5. Filter contours on size and aspect ratio.
6. Calculate minimum bounding box. 

The shape of the fiducial ensures that the center of the box is the center of the fiducial and the orientation of the box is the orientation of the fiducial.
